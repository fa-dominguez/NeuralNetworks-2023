```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

```python
# Load data into pandas
df_data = pd.read_excel('./Deep Learning_ Mastering Neural Networks - Module 1 Assignment Spreadsheet.xlsx')

# View the first 5 components
df_data.head()
```

```python
# Info of dataset
df_data.info()
```

```python
# Linear Regression (OLS)

# Store x-values
x = df_data[['X']]

# Store y-values
y = df_data['Y']

# Create linear regression model
reg = LinearRegression().fit(x, y)

# Store coefficient
a = round(reg.coef_[0], 3)

# Store intercept
b = round(reg.intercept_, 4)

# Linear regression score
score = round(reg.score(x,y), 5)

# print coefficient and intercept
print("Linear Regression x_0 coefficient: ", a)
print("Linear Regression intercept:       ", b)
print("Linear Regression Score:           ",  score)

# Visualize linear regression
x_plot = np.linspace(-5, 10, 10)

# Define line equation
y_plot = a*x_plot + b

# Plot equation
plt.plot(x_plot, y_plot, label = f'y = {a}*x + {b}')
plt.scatter(x, y)
plt.title(f'Linear Regression')
plt.legend()
plt.show()
```

```python
# Linear Regression (Ridge)

# Store x-values
x = df_data[['X']]

# Store y-values
y = df_data['Y']

# Create and fit Ridge model
ridge_model = Ridge().fit(x, y)

a = round(ridge_model.coef_[0], 3)

# Store intercept
b = round(ridge_model.intercept_, 5)

# Store score
score = round(ridge_model.score(x,y), 4)

# print coefficient and intercept
print("Ridge Regression x_0 coefficient: ", a)
print("Ridge Regression intercept:       ", b)
print("Ridge Regression Score:           ",  score)

# Visualize linear regression
x_plot = np.linspace(-5, 10, 10)

# Define line equation
y_plot = a*x_plot + b

# Plot equation
plt.plot(x_plot, y_plot, label = f'y = {a}*x + {b}')
plt.scatter(x, y)
plt.title(f'Ridge Regression')
plt.legend()
plt.show()

```

<h1> Comments</h1><br>
<div align = 'justify'>
This code runs two linear regression models. The first one is commonly referred to as <b>Linear Regreassion</b> and is one of the most popular models. Nonetheless, this model is optimized using the <b>Ordinary Least Squares error method</b>, which minimizes the square error over the data points. This model returns the same result as the one presented in the Excel file, as Excel utilize by default the OLS regression model.<br><br>

The second model also performs a linear regression but uses the <b>Ridge optimization method</b>. This method regularizes the model by adding a penalty to large values of models coefficients (thetas or weights). This model aims to reduce the coefficient as close as possible to zero, helping to address the problem of overfitting a model.
